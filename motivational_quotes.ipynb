```python
def motivational_quotes(motiv):
    quotes= [i.text.strip() for i in motiv.find_all('div',class_='quoteText') ]
    author=[i.text.strip() for i in motiv.find_all('span',class_='authorOrTitle')]
    likes=[int(i.text.strip().split()[0]) for i in motiv.find_all('a',class_='smallText')]
    tages = [i.text.split()[1:] for i in[i for i in motiv.find_all('div',class_='greyText smallText left')]]
    return pd.DataFrame({'Author':author,'Quotes':quotes,'Likes':likes,'Tages':tages})
```

```python
motivational_quotes(motiv)
```
